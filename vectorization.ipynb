```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
```

```python
df = pd.read_csv('data/processed_resume_match.csv')

# Vectorizer for the combined classification text
combined_tfidf = TfidfVectorizer(max_features=2000, ngram_range=(1, 2))
X_classification = combined_tfidf.fit_transform(df['input_text'].fillna("")).toarray()
```

```python
# Save vectorizers and arrays
joblib.dump(combined_tfidf, 'models/combined_tfidf.pkl')

np.save('models/X_class.npy', X_classification)
np.save('models/y_target.npy', df['result'].values)
print("Notebook 2 complete. Feature spaces created.")
```
